#### Separate digitals in 'Energy' column

```python
vals = df['En'].to_csv(header=None, index=False).strip('\n').split('\n')

import re

digit = []
for val in vals:
    found = re.findall(r"[0-9./]",val)
    if found:
        digit.append(found)
        digits = []
    for i in range(len(digit)):
        found = ''.join(digit[i])
        if found:
            digits.append(found)
            
len(digits)
```
